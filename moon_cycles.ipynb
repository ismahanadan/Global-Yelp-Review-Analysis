```python
url = "http://api.worldweatheronline.com/premium/v1/astronomy.ashx?"
q = "Las Vegas"

blackfridays = ["2009-11-27", "2010-11-26", "2011-11-25", "2012-11-23", "2013-11-29", "2014-11-28"]
christmas_eves = ["2009-12-24", "2010-12-24", "2011-12-24", "2012-12-24", "2013-12-24", "2014-12-24"]
mothers_days = ["2009-05-10", "2010-05-09", "2011-05-08", "2012-05-13", "2013-05-12", "2014-05-11"]
uk_mothers_day = ["2009-03-22", "2010-03-14", "2011-04-03", "2012-03-18", "2013-03-10", "2014-03-30"]
four_20s = ["2009-04-20", "2010-04-20", "2011-04-20", "2012-04-20", "2013-04-20", "2014-04-20"]
valentines_day = ["2009-02-14", "2010-02-14", "2011-02-14", "2012-02-14", "2013-02-14", "2014-02-14"]

dates = blackfridays + christmas_eves + mothers_days + uk_mothers_day + four_20s + valentines_day
```

```python
date_list = []
moon_cycle = []
```

```python
for date in dates:
    
    query_url = url + "q=" + q + "&date=" + str(date) + "&format=json" + "&key=" + api_key
    moon_data = requests.get(query_url).json()      
    
    date_list.append(date)
    moon_cycle.append(moon_data['data']['time_zone'][0]['moon_phase'])
```

```python
moon_dict = {
    "Date" : date_list,
    "Moon Phase": moon_cycle
}
```

```python
moon_df = pd.DataFrame(moon_dict)
```

```python
#read business file and create dataframe 

business_file = "../Resources/yelp_academic_dataset_business.csv"
business_df = pd.read_csv(business_file)
business_df.head()
```

```python
#read review csv file and create dataframe 

review_file = "../Resources/yelp_academic_dataset_review.csv"
review_df = pd.read_csv(review_file)
review_df.head()
```

```python
#from business dataframe create a filtered dataframe from specific columns 

new_business_df = business_df[['business_id', 'state', 'city', 'categories']].copy()
new_business_df.head()
```

```python
#drop Nan values in data
new_business_df = new_business_df.dropna(subset=['categories'])
new_business_df.head(20)
```

```python
new_business_df = new_business_df[new_business_df['categories'].str.lower().str.contains("restaurants")]
```

```python
#describe business dataframe

new_business_df.describe()
```

```python
#from review dataframe create a filtered dataframe from specific columns 

new_review_df = review_df[['business_id', 'user_id', 'review_id','stars', 'date']].copy()
new_review_df.head()
```

```python
#merge dataframes

yelp_df = pd.merge(new_business_df, new_review_df, on="business_id")
yelp_df
```

```python
#clean the data by dropping duplicates and setting the index

yelp_df.drop_duplicates("review_id", inplace=True)
```

```python
yelp_df.set_index("user_id", inplace=True)

yelp_df.head()
```
